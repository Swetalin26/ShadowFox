Python Visualization Libraries: Matplotlib vs Seaborn

## 1. Library Overview

### ðŸ”¹ Matplotlib
Matplotlib is a low-level 2D plotting library that allows precise control over every aspect of a figure. It is widely used for creating static, animated, and interactive plots in Python.

- Key Features:
  - Highly customizable
  - Supports various output formats
  - Works well with NumPy and Pandas

Use Case: When you need full control over plot elements.

---

###  ðŸ”¹ Seaborn
Seaborn is a high-level interface built on top of Matplotlib that makes it easier to create beautiful, statistical graphics.

- Key Features:
  - Built-in themes and color palettes
  - Simplified syntax
  - Great for statistical plots

Use Case: When you want quick and attractive plots with minimal code.


**GRAPH TYPES IN MATPLOTLIB:**

**1.LINE PLOT:**


```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

**2.BAR CHART**

```python
x = ['Apple', 'Banana', 'Mango']
y = [10, 15, 7]

plt.bar(x, y)
plt.title("Bar Chart")
plt.show()
```

**SCATTER PLOT:**

```python
x = [1, 2, 3, 4]
y = [30, 50, 70, 90]

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()
```

**HISTOGRAM:**

```python
marks = [60, 70, 80, 90, 85, 75, 65, 95, 70, 80]

plt.hist(marks)
plt.title("Histogram")
plt.show()
```

**PIE CHART:**

```python
labels = ['Food', 'Rent', 'Shopping']
sizes = [40, 30, 30]

plt.pie(sizes, labels=labels)
plt.title("Pie Chart")
plt.show()
```

**GRAPH FOR SEABORN:**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample dataset
np.random.seed(42)
df = pd.DataFrame({
    "Category": np.random.choice(['A', 'B', 'C'], size=100),
    "Value": np.random.normal(50, 10, size=100),
    "Score": np.random.randint(20, 100, size=100),
    "Gender": np.random.choice(['Male', 'Female'], size=100)
})

# 1. Line Plot
sns.lineplot(data=df.sort_values("Score"), x="Score", y="Value")
plt.title("Line Plot")
plt.show()

# 2. Bar Plot
sns.barplot(x="Category", y="Value", data=df)
plt.title("Bar Plot")
plt.show()

# 3. Count Plot
sns.countplot(x="Category", data=df)
plt.title("Count Plot")
plt.show()

# 4. Box Plot
sns.boxplot(x="Category", y="Value", data=df)
plt.title("Box Plot")
plt.show()

# 5. Violin Plot
sns.violinplot(x="Category", y="Value", data=df)
plt.title("Violin Plot")
plt.show()

# 6. Strip Plot
sns.stripplot(x="Category", y="Value", data=df, jitter=True)
plt.title("Strip Plot")
plt.show()

# 7. Swarm Plot
sns.swarmplot(x="Category", y="Value", data=df, size=3)
plt.title("Swarm Plot")
plt.show()

# 8. Scatter Plot
sns.scatterplot(x="Score", y="Value", hue="Gender", data=df)
plt.title("Scatter Plot")
plt.show()

# 9. Histogram
sns.histplot(data=df, x="Value", bins=10)
plt.title("Histogram")
plt.show()

# 10. KDE Plot
sns.kdeplot(data=df, x="Value", fill=True)
plt.title("KDE Plot")
plt.show()

# 11. Heatmap
corr = df[["Value", "Score"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Heatmap")
plt.show()
```

              
 **Matplotlib**                                         
 Ease of Use - Lower-level, requires more code to build complex plots 
 Customization - Highly customizable (colors, axes, styles, etc.)       
 Interactivity - Static by default; can integrate with `mplcursors`, interactive backends  
Performance (Large Data) - Efficient for large datasets (if coded carefully)    
Learning Curve - Steeper (more manual control, more verbose)            
Integration - Works with NumPy, Pandas, Jupyter                     
Visual Aesthetics - Plain by default, but fully stylable                            |


**SEABORN**
Ease of Use - High-level API, easier for beginners and faster to plot
Customization -  Limited customization (built-in styles, some overrides)        
Interactivity - Primarily static; depends on Matplotlib's interactivity  
Performance (Large Data) -Slower with large categorical data; more overhead   
Learning Curve - Easier (sensible defaults, readable syntax)              
Integration -  Built on top of Matplotlib and integrates with Pandas                     
Visual Aesthetics -  Beautiful default themes and color palettes
